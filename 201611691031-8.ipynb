```python
def total_(m,n,k):
    total=0
    i=m
    while i<=n:
        total=total+i
        i=i+k
    return total

m=int(input('请输入一个整数'))
n=int(input('请输入一个整数'))
k=int(input('请输入间隔数'))
print('最后的和是',total_(m,n,k))


```
